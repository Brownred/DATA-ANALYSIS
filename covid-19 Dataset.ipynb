### Get the total cases, total deaths per continent

### Break the dataset into smaller subsets
##### Filter specific rows from a DataFrame
iminterested in the data with total cases greater than 500,000

```python
High_total_cases = Covid[Covid["total_cases"] >500000]
High_total_cases.shape
# High_total_cases.describe(include="all")
```

```python
High_total_cases.head(5)
```

I want to work with the data for which the continent is known. Do away with the data in continent column that is not known(NaN)

```python
# to check the total number af null values in continent colum
High_total_cases["continent"].isna().sum()


#Case_per_continent = High_total_cases[High_total_cases["continent"].notna()]


# alternative is using the drop method
Case_per_continent = High_total_cases.dropna(subset=["continent"])
```

```python
# List The continent name
Continents_unique = Case_per_continent.continent.unique()
list(Continents_unique)
```

```python
# Grouping the data by continents and summing them
Case_per_continent = Case_per_continent.groupby(['continent'])[["total_cases","total_deaths","total_vaccinations"]].sum()
Case_per_continent
```

```python
# Continent shouldnt be an index
Case_per_continent.reset_index(level=0,inplace=True)
Case_per_continent
```
